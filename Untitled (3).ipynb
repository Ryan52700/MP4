# Traitement de l'image avec PIL

## Exercice 1

**1)** La définition d'une image est le nombre de pixel total de l'image. Elle est écrite de cette manière : 1200/800 où 1200 correspond au nombre de pixels sur la longueur et 800 sur la largeur.

**2)** La résolution d'un image correspond au nombre de pixel par pouce (PPP). C'est cette résolution qui va noius indiquer la qualité de l'image et non pas la définition.

**3)** La profondeur de couleur est exprimé en bpp (bit par pixel) et représente le nombre de bits utilisés dans un pixel pour représenter la couleur. Pour obtenir une grande échelle de nuance, il faudra avoir un bpp élevé. Il existe 3 grands types de profondeur de couleur :

- 8-bit Color (couleur)
- 15/16 bit : Highcolor (haute couleur)
- 24/32 bit : Truecolor (vraie couleur)

## Exercice 4

```python
def definition(image, PixelLongueur, PixelLargeur):
    result = PixelLongueur / image.width
    result2 = PixelLargeur / image.height
    
    print("La définition est {}/{}".format(result, result2))
```

```python
img1 = Image.open("iguane.jpg")
definition(img1, 3000, 4500)
```

## Exercice 5

```python
def symetries(image):
    
    choix = int(input("Symétrie horizontale : 1, Symétrie verticale : 2. Faites votre choix : "))
    img2 = Image.new("RGB",(image.width,image.height))
    
    if choix == 1:
        for h in range(image.height):
            for i in range(image.width):
                couleur = image.getpixel((i,h))
                R,V,B = couleur
                img2.putpixel((img2.width - 1 - i, h),(R,V,B))
                
    if choix == 2:
        for i in range(image.width):
            for h in range(image.height):
                couleur = image.getpixel((i,h))
                R,V,B = couleur
                img2.putpixel((i, img2.height - 1 - h),(R,V,B))
    display(img2)
    
```

```python
img1 = Image.open("iguane.jpg")
symetries(img1)
```

## Exercice 6

```python
def niveauDeGris(image):
    
    img2 = Image.new("RGB",(image.width,image.height))
    
    for i in range(image.width):
        for h in range(image.height):
            
            couleur = image.getpixel((i,h))
            R,V,B = couleur
            G = (R + V + B) // 3
            img2.putpixel((i,h),(G,G,G))   
            
    display(img2)
```

```python
niveauDeGris(img1)
```

## Exercice 7

```python
def negatif(image):
    
    img2 = Image.new("RGB",(image.width,image.height))
    
    for i in range(image.width):
        for h in range(image.height):
            
            couleur = image.getpixel((i,h))
            R,V,B = couleur
            img2.putpixel((i,h),(255 - R,255 - V,255 - B))   
    
    display(img2)
```

```python
negatif(img1)
```
